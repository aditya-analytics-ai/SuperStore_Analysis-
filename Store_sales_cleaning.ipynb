```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading the dataset
df=pd.read_csv("Sample_Superstore.csv")
```

```python
#Fixing Column Name 
df.columns=df.columns.str.replace(" ","_")
df.columns
```

```python
#Fixing Column Data type
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='mixed', errors='coerce')
```

```python
df['Order_Date'].isnull().sum()
```

```python
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], format='mixed', errors='coerce')
```

```python
df['Ship_Date'].isnull().sum()
```

```python
df['Postal_Code'] = df['Postal_Code'].astype(str).str.zfill(5)
```

```python
df['Product_Name'] = df['Product_Name'].str.strip()
```

```python
df.duplicated(subset=['Order_ID', 'Product_ID']).sum()
```

```python
loss_rows = df[df['Profit'] < 0]
loss_rows
```

```python
loss_rows[['Discount','Sales','Profit']].describe()
```

```python
df.drop(columns=['Country'], inplace=True)
```

```python
df.drop(columns=['Row_ID'], inplace=True)
```

```python
df = df.drop_duplicates(subset=['Order_ID', 'Product_ID'], keep='first')
```

```python
df['Days_to_Ship'] = (df['Ship_Date'] - df['Order_Date']).dt.days
```

```python
df['Order_Year']    = df['Order_Date'].dt.year       
df['Order_Month']   = df['Order_Date'].dt.month
df['Order_Quarter'] = df['Order_Date'].dt.quarter 
```

```python
df
```

```python
df.to_csv("Superstore_Cleaned.csv", index=False)
```
